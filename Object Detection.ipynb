```python
import cv2
import time
import imutils
```

```python
# It is used for resize the image
img = cv2.imread("AI/i1.PNG")
resizeImg = imutils.resize(img,width=30)
cv2.imwrite("AI/resizeImg.PNG",resizeImg)
```

```python
# it is used for blur image

img = cv2.imread("AI/i1.PNG")
gaussianimg = cv2.GaussianBlur(img,(25,25),0)
cv2.imwrite("AI/gaussianImg.PNG",gaussianimg)
```

```python
# threshold image : it is used for to fill the black colour in picture
img = cv2.imread("AI/i1.PNG")
grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
thresImg = cv2.threshold(grayImg,220,225,cv2.THRESH_BINARY)[1]
cv2.imwrite("AI/thresholdImg.PNG",thresImg)
```

```python
# To take screenshot of picture using laptopcamera

cam = cv2.VideoCapture(0)
time.sleep(1)


_, img = cam.read()
cv2.imwrite("AI/cameraImg.jpg",img)
cam.release()
```

```python
cam = cv2.VideoCapture(0)
time.sleep(1)

while True:
    _, img = cam.read()
    cv2.imshow("AI/camerafeed",img)
    key = cv2.waitKey(1) & 0xFF # It will return in 8 bit 
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture(0)
time.sleep(1)

firstFrame = None
area = 400

while True:
    _, img = cam.read()
    text = "Normal"
    img = imutils.resize(img,width=500)
    grayImg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    gaussianImg = cv2.GaussianBlur(grayImg, (21,21),0)
    if firstFrame is None:
        firstFrame = gaussianImg
        continue
    imgDiff = cv2.absdiff(firstFrame, grayImg)
    threshImg = cv2.threshold(imgDiff,25,255,cv2.THRESH_BINARY)[1]
    threshImg = cv2.dilate(threshImg,None,iterations=2)    # dilate is used to fill the hole after completed the first threshold image.
    # contours is connected to the neighberhood pictures
    cnts = cv2.findContours(threshImg.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    cnts = imutils.grab_contours(cnts)
    for c in cnts:
        if cv2.contourArea(c) < area:
            continue
        (x,y,w,h) = cv2.boundingRect(c)
        cv2.rectangle(img, (x,y), (x + w,y + h),(0,255,0),2)
        text = "Object detected"
        print(text)
    cv2.putText(img,text,(10,20),cv2.FONT_HERSHEY_SIMPLEX , 0.5, (0, 0 , 255),2)
    
    cv2.imshow("cameraFeed",img)
    key = cv2.waitKey(1) & 0xFF # It will return in 8 bit 
    if key == ord("q"):
        break
cam.release()
cv2.destroyAllWindows()
```
